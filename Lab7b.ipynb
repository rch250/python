```python
import pyodbc
import pandas as pd
import seaborn as sns
```

```python
connection = pyodbc.connect(
    server = "cisdbss.pcc.edu",
    user = "275student",
    password = "275student",
    database = "NAMES",
    driver = "{ODBC Driver 17 for SQL Server}"
)

connection
```

```python
df = pd.read_sql_query("""
SELECT Name, Year, NameCount, Gender 
FROM all_data
WHERE Name = 'Robert'
""", connection)

df
```

```python
df.info()
```

### Relational Plot: Line

```python
g = sns.relplot(data=df, kind='line',
            x='Year', y='NameCount', hue='Gender')
g
```

```python
g.savefig('NamesLineChart.png')
g.savefig('NamesLineChart.svg')
```

### Distribution: Histogram

```python
g = sns.displot(data=df, kind='hist', x='Year', y='NameCount', hue='Gender')
```

```python
g.savefig('NamesHistChart.png')
g.savefig('NamesHistChart.svg')
```

### Category: Bar

```python
g = sns.catplot(data=df, 
            kind='bar', x='Year', y='NameCount', hue='Gender', orient='h')
g
```

```python
g.savefig('NamesBarChart.png')
g.savefig('NamesBarChart.svg')
```

### Subplots: Lines

```python
g=sns.relplot(data=df, 
            kind='line', x='Year', y='NameCount', hue='Gender', 
            height=4, aspect=1.25, 
            col='Gender', col_wrap=2, legend=False)
g
```

```python
g.savefig('NamesLinesubChart.png')
g.savefig('NamesLinessubChart.svg')
```

### Subplots: RelScatter

```python
g=sns.relplot(data=df, 
            kind='scatter', x='Year', y='NameCount', hue='Gender', 
            height=4, aspect=1.25, 
            col='Gender', col_wrap=2, legend=False)
g
```

```python
g.savefig('NamescattersubChart.png')
g.savefig('NamescattersubChart.svg')
```
